##Recodificação dos dados

Um dos grandes problemas dos arquivos *DBF* é a codificação das variáveis. Simplesmente abrir o arquivo não traz informações claras, pois acabamos tendo que recorrer ao Dicionário de Dados daquele arquivo. Com o *Python*, conseguimos recodificar as colunas, escrevendo o dicionário de dados para cada uma. É um trabalho que leva algum tempo, porém, precisamos fazê-lo somente uma vez e depois já estará pronto.


Avaliando o dicionário de dados, percebemos que muitas variáveis apresentam apenas o **SIM** e o **NÃO**. Para essas, faremos apenas um mini dicionário e aplicaremos a função **REPLACE**, que vai substituir os valores encontrados na tabela por aqueles vistos no dicionário.

Utilizaremos uma coluna como exemplo, e depois aplicamos para as demais.

```python
dados['MCLI_LOCAL'] = dados['MCLI_LOCAL'].replace({1:'Sim',
                                                   2:'Não',
                                                   9:'Ignorado'})
dados.head()
```

Agora, criaremos uma lista com as colunas que possuem codificação de **SIM/NÃO**, e aplicaremos uma das funções mais úteis do *Python*, o loop de repetição chamado de **FOR**.

Essa função nos permite aplicar diversas vezes a mesma operação, alterando apenas um parâmetro.

```python
colunas_sim_nao = ['MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA',
       'CLI_EQUIMO', 'CLI_NECROS', 'CLI_LOCAL_', 'MCLI_SIST',
       'CLI_NEURO', 'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL',
       'CLI_OUTR_2','CON_SOROTE','COM_LOC',
       'COM_SECUND', 'COM_NECROS', 'COM_COMPOR', 'COM_DEFICT', 'COM_APUTAC',
       'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC', 'COM_CHOQUE',
       'DOENCA_TRA']

for coluna in colunas_sim_nao:
  dados[coluna] = dados[coluna].replace({1:'Sim',
                                         2:'Não',
                                         9:'Ignorado'})

dados
```

A coluna de **IDADE** é um pouco diferente. O SINAN utiliza uma decodificação neste formato:
A composição da variável
obedece o seguinte critério: primeiro dígito:
1. Hora
2. Dia
3. Mês
4. Ano

*Exemplos*:
 * 3009 – nove meses
 * 4018 – dezoito anos

Assim, como apenas nos interessa a idade em anos, diminuiremos todos os números por 4000. Caso o resultado for negativo, deixamos como zero.

```python
idades = []
for i in dados['NU_IDADE_N']:
    if i > 4000:
      idade = i - 4000
    
    elif 4000 > i:
      idade = 0

    idades.append(idade)

dados['NU_IDADE_N'] = idades

dados
```

##Código do IBGE para nome de município

Da mesma maneira que utilizamos um dicionário e a função *replace* para alterar elementos no DataFrame, faremos o mesmo para transformar os códigos de IBGE em nomes de municípios, utilizando a função *map*.

Primeiramente, precisamos de uma tabela onde esta conversão já esteja pronta. Por sorte, aquele que vos escreve já o fez e colocou um link no *Github* para baixá-lo. ✌

```python
municipios = pd.read_csv('https://raw.githubusercontent.com/decao88/csv/master/populacao%20ibge%206%20municipio%20br.csv', sep=';')
municipios.head()
```

Porém, para convertê-lo em dicionário, precisamos realizar algumas alterações, como transformar o código do IBGE em tipo *texto* (*string*), deixar apenas duas colunas (IBGE6 e Municipio), e setar a coluna IBGE6 como index.

```python
municipios = municipios[['IBGE6',	'Municipio']]
municipios = municipios.set_index('IBGE6')
```

Agora, basta utilizar a função **to_dict()** e teremos o nosso dicionário

```python
dicionario_municipios = municipios.to_dict()['Municipio']
```

Para transformar os dados, agora vamos utilizar a função *map*. Pode surgir a dúvida: **"Por que aqui utilizamos o map e antes utilizamos o replace? Existe diferença?"**

Respondendo: O map é uma função que, caso o valor não esteja no dicionário, ele se transforma em *NaN*, fazendo com que percamos aquele dado, porém, ela é 60 vezes mais rápida do que o replace, então ganhamos tempo quando o dicionário possui mais de 5000 valores.

Assim, convertemos as duas colunas que onde temos os códigos do IBGE (ID_MUNICIP, ID_MN_RESI e ANT_MUNIC_).

```python
dados['ID_MUNICIP_nome'] = dados['ID_MUNICIP'].map(dicionario_municipios)
dados['ID_MN_RESI_nome'] = dados['ID_MN_RESI'].map(dicionario_municipios)
dados['ANT_MUNIC__nome'] = dados['ANT_MUNIC_'].map(dicionario_municipios)


dados.head()
```

Da mesma maneira, é possível alterar a codificação da OCUPAÇÃO

```python
ocupacao = pd.read_csv('https://raw.githubusercontent.com/andrejarenkow/csv/master/CBO2002%20-%20Ocupacao.csv')

ocupacao = ocupacao.set_index('CODIGO')
dicionario_ocupacao = ocupacao.to_dict()['TITULO']
dados['ID_OCUPA_N'] = dados['ID_OCUPA_N'].map(dicionario_ocupacao)
```

#Aplicação de filtros

Uma coisa que muitas vezes fazemos ao analisar um banco de dados, é um filtro. Digamos que eu queira verificar SOMENTE os dados de acidentes ocorridos em Santa Catarina.

A primeira coisa a fazer é declarar o filtro, e depois vamos aplicá-lo.

```python
filtro_sc = dados_final['UF do acidente'] == 'SC'

dados_final_sc = dados_final[filtro_sc].reset_index(drop=True)
dados_final_sc
```

E se quisermos fazer um DataFrame com mais de um parâmetro de filtro? Algumas maneiras de fazer:
* 1 Fazer um filtro seguido do outro - desvantagem é que não podemos utilizar parâmetros dentro da mesma coluna, como por exemplo selecionar mais de um UF. 

```python
filtro_sc = dados_final['UF do acidente'] == 'SC'
filtro_serpente = dados_final['Tipo de acidente'] == 'Serpente'

dados_final_sc_serpente = dados_final[filtro_sc][filtro_serpente]
dados_final_sc_serpente
```

* 2 - Filtros composto: é o melhor jeito de fazer, podendo colocar filtros com E (&) / OU ( | )

```python
#criando filtros
filtro_sc = dados_final['UF do acidente'] == 'SC'
filtro_rs = dados_final['UF do acidente'] == 'RS'
filtro_serpente = dados_final['Tipo de acidente'] == 'Serpente'

#juntando filtros
filtro_composto = (filtro_sc | filtro_rs)&filtro_serpente

#aplicando filtro composto
dados_final_sc_rs_serpente = dados_final[filtro_composto]
dados_final_sc_rs_serpente
```

* 3 - Filtro composto para a mesma coluna, criando uma lista: quando queremos fazer um filtro que é sempre na mesma coluna, mas temos vários itens. Aplica-se o filtro utilizando a função *isin*

```python
#criando lista
lista_uf = ['BA','DF','PE','SC','RS']

#definindo o filtro
filtro_lista_uf = dados_final['UF do acidente'].isin(lista_uf)

#aplicando o filtro
dados_final_lista = dados_final[filtro_lista_uf]
dados_final_lista
```

##PROCV do Python - juntando duas tabelas
Uma das funções mais utilizadas no Excel é o **PROCV**, que busca o valor em uma coluna e acrescenta na outra. No Python, é utilizada a função do pandas **merge**. Vamos utilizar esta função para verificar a incidência dos acidentes em cada UF, dividindo o número de casos pelo número de habitantes da UF. [Material de referência](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html)

O primeiro passo é gerarmos uma tabela com o número de acidentes por UF. Já fizemos ela anteriormente com o **value_counts()**, agora vamos transformá-la em DataFrame.

Após, vamos renomear as colunas.

```python
acidentes_uf_ocorrencia = pd.DataFrame(dados_final['UF do acidente'].value_counts()).reset_index()
acidentes_uf_ocorrencia.columns = ['UF',	'Acidentes']
acidentes_uf_ocorrencia
```

E onde vamos achar uma tabela do número de habitantes por UF? 🥸

Eu gosto de ir no site do [DataSUS/TabNET/Demográficas e Socioeconômicas/População Residente](https://datasus.saude.gov.br/populacao-residente), pela facilidade. Mas é livre a qualquer um buscar esta informação.

Felizmente, o prof já providenciou o arquivo anteriormente kkkk

```python
tabela_uf
```

Agora, utilizamos o comando **MERGE** do Pandas, o qual irá juntas as duas tabelas.

```python
acidentes_uf_pop = pd.merge(acidentes_uf_ocorrencia, tabela_uf, on='UF')
acidentes_uf_pop
```

Agora, realizamos uma operação de divisão entre as colunas *Acidentes* e a coluna *Populacao*

```python
acidentes_uf_pop['Incidencia/100.000 hab'] = acidentes_uf_pop['Acidentes']/acidentes_uf_pop['Populacao']*100000
acidentes_uf_pop
```

Se quisermos ordenar o DataFrame por incidência, devemos utilizar a função *sort_values*

```python
acidentes_uf_pop.sort_values('Incidencia/100.000 hab', ascending=False)
```

##Plotly Express

Uma biblioteca de visualização de dados que está ganhando popularidade atualmente é a [Plotly Express](https://plotly.com/python/plotly-express/#gallery), uma vez que já traz gráficos bastante robustos, e que parecem até um *Dashboar*.

Podemos inclusive fazer mapas com ela!

Primeiramente, vamos importar o arquivo GeoJson,que sera a nossa base para o mapa.

```python
import json
import urllib.request
import plotly.express as px

with urllib.request.urlopen("https://github.com/andrejarenkow/geodata/blob/main/municipios_rs_CRS/rs_municipios_crs.json?raw=true") as url:
    rs_municipios = json.loads(url.read().decode())
```

```python
rs_municipios["features"][0]
```

Como o arquivo é para os municípios do RS, vamos criar um filtro que mostra apenas acidentes ocorridos no Estado.

Após, vamos gerar uma tabela de contagem de valores por município de ocorrência.

```python
dados_final_rs = dados_final[dados_final['UF do acidente']=='RS']
tabela_dinamica = pd.pivot_table(data=dados_final_rs, index=['Código Município de ocorrência do acidente','Município de ocorrência do acidente'], values='UF do acidente', aggfunc='count').reset_index()
tabela_dinamica.columns = ['IBGE6',	'Município de ocorrência do acidente',	'Número de acidentes']
tabela_dinamica
```

Agora vem a parte onde devemos juntar tudo e gerar o nosso mapa!

```python
fig = px.choropleth_mapbox(tabela_dinamica, geojson=rs_municipios, color='Número de acidentes',
                           locations="IBGE6", featureidkey="properties.IBGE6",
                           center={"lat": -29.410455452143456, "lon": -53.10333642259305}, 
                           mapbox_style="open-street-map", zoom=5, hover_data=['Município de ocorrência do acidente'])
fig.show()
```

Como já existe um mapa para a Regional de Chapecó, em Santa Catarina, vamos fazer este mapa também!

```python
with urllib.request.urlopen("https://raw.githubusercontent.com/andrejarenkow/geodata/main/Regional_Chapeco_SC/regional_chapeco_sc.geojson") as url:
    sc_municipios = json.loads(url.read().decode())

dados_final_sc = dados_final[dados_final['UF do acidente']=='SC']
tabela_dinamica = pd.pivot_table(data=dados_final_sc, index=['Código Município de ocorrência do acidente','Município de ocorrência do acidente'], values='UF do acidente', aggfunc='count').reset_index()
tabela_dinamica.columns = ['IBGE6',	'Município de ocorrência do acidente',	'Número de acidentes']

fig = px.choropleth_mapbox(tabela_dinamica, geojson=sc_municipios, color='Número de acidentes',
                           locations="IBGE6", featureidkey="properties.IBGE6",
                           center={"lat": -26.79887472771733, "lon": -52.792938492227925},
                           mapbox_style="open-street-map", zoom=7, hover_data=['Município de ocorrência do acidente'])
fig.show()
```

##Gráfico de barras

Vamos explorar outros tipos de gráficos também. Para isso, utilizaremos uma tabela já vista anteriormente, sobre o tipo de acidente por UF, e resetaremos o índice dela, para que a UF do acidente se transforme em uma coluna.


```python
tabela_animal_uf.reset_index(inplace=True)
tabela_animal_uf
```

Para que o gráfico fique adequado, precisamos reestruturar a nossa tabela, deixando-a com apenas 3 colunas:
* UF do acidente
* Tipo do acidente
* Número de acidentes

Assim, vamos utilizar uma função do Pandas chamada [**melt**](https://pandas.pydata.org/docs/reference/api/pandas.melt.html), a qual vai "derreter" a tabela e deixá-la com esta nova estrutura.

```python
tabela_melt = pd.melt(tabela_animal_uf, id_vars=['UF do acidente'], value_vars=['Abelha',	'Aranha',	'Escorpião',	'Ignorado',	'Lagarta',	'Outros',	'Serpente'],value_name = 'Acidentes')
tabela_melt
```

Agora o gráfico pode ser contruído.

```python
fig = px.bar(tabela_melt, x="UF do acidente", y="Acidentes", color="Tipo de acidente")
fig.show()
```

Vamos repetir o processo, mas avaliando os números absolutos, e não mais os normalizados.

```python
tabela_animal_uf_absoluto = (pd.crosstab(dados_final['UF do acidente'], dados_final['Tipo de acidente'])).reset_index()
tabela_melt_absoluto = pd.melt(tabela_animal_uf_absoluto, id_vars=['UF do acidente'], value_vars=['Abelha',	'Aranha',	'Escorpião',	'Ignorado',	'Lagarta',	'Outros',	'Serpente'], value_name = 'Acidentes')
fig = px.bar(tabela_melt_absoluto, x="UF do acidente", y="Acidentes", color="Tipo de acidente")
fig.show()
```
